```python
import os
import gymnasium as gym
from stable_baselines3 import A2C
from stable_baselines3.common.vec_env import DummyVecEnv, VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy

from stable_baselines3.common.env_util import make_atari_env
```

```python
import ale_py

gym.register_envs(ale_py)
```

```python
env = make_atari_env("Breakout-v4", n_envs=4, seed=0)
```

```python
env = VecFrameStack(env, n_stack=4)
```

```python
log_path = os.path.abspath('/home/dingus_ferry/Development/RL/practice/logs/')
model_path = os.path.abspath('/home/dingus_ferry/Development/RL/practice/models/A2C_Breakoutv4_Model')
```

```python
model = A2C("CnnPolicy", env, verbose=1, tensorboard_log=log_path, device='cuda')
```

```python
model.learn(total_timesteps=1_000_000)
```

```python
model.save(model_path)
```

```python
evaluate_policy(model, env, n_eval_episodes=5, render=True)
```

```python
del model
```

```python
model = A2C.load(model_path)
```

```python
env = make_atari_env('Breakout-v4', n_envs=1, seed=0)
env = VecFrameStack(env, n_stack=4)
```
