# PPO

```python
import os
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
environment_name = "CartPole-v1"
env = gym.make(environment_name, render_mode="human")
```

```python
episodes = 5
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0 
    
    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, trunc, info = env.step(action)
        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
env.close()
```

```python
log_path = os.path.abspath('/home/dingus_ferry/Development/RL/practice/logs/')
```

```python
env = DummyVecEnv([lambda: env])
```

```python
model = PPO('MlpPolicy', env, verbose = 1)
```

```python
model.learn(total_timesteps=100000)
```

```python
model_path = os.path.abspath('/home/dingus_ferry/Development/RL/practice/models/PPO_CartPolev1_Model')
```

```python
model.save(model_path)
```

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
```

```python
env.close()
```

# DQN

```python
environment_name = "CartPole-v1"
env = make_vec_env(environment_name, n_envs=4)
```
